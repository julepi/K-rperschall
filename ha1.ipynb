```python
from pylab import *
import sympy as sp


#Zentrifugalkraft F=mw^2r
#Drehzahl n=[0,1200]rpm, rpm=1/60s, n=[0,20]1/s 
n=arange(0,20.001,0.01)
#Kreisfrequenz
w=2*pi*n
mb=7 #7 kg Beladung
r=0.3 #Radius Waschmaschinentrommel
Fz=mb*w**2*r
plot(w,Fz)
xlabel('$\omega$ in Hz')
ylabel('$F_z$ in N')
```

```python

mg=87
Fg=mg*9.81
E=array([0.03,0.3,1])
E=10**6*E
A=0.16E-4
l=0.025
sg=E*A/l
w0=sqrt(sg/mg)
eta=array([0.003,0.23,1])
for i in range(3):
    for j in range(3):
        cg[i]=eta[j]*sg[i]/w0[i]
        
cg
```

```python
import sympy as sy
om,m,s,c,eta=sy.symbols('omega m s c eta')
H=sy.Function('H')
a=sy.Function('a')
p=sy.Function('p')
H=1/(1j*c*om-m*om**2+s)
H
```

```python
a=sy.Abs(H).simplify()
a
```

```python
p=sy.arg(H).simplify()
p
```

```python
import numpy as np # numeric math
import matplotlib.pyplot as pl
amp=sy.lambdify([c,om,m,s],a)
ph=sy.lambdify([c,om,m,s],p)

for i in range(3):
    for j in range(3):
        figure(1)
        pl.loglog(w,amp(cg[i],w,mg,sg[j]),label='Feder %d, Dmpfer %d'%(j+1,i+1))
        xlabel('$\omega$ in Hz')
        ylabel('abs(H)')
        pl.title('Amplitudengang')
        legend()
        figure(2)
        pl.plot(w,ph(cg[i],w,mg,sg[j]),label='Feder %d, Dmpfer %d'%(j+1,i+1))
        pl.xscale('log')
        xlabel('$\omega$ in Hz')
        ylabel('arg(H)')
        pl.title('Phasengang')
        legend()
```
